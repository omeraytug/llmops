# Explore chat in PydanticAI

```python
from pydantic_ai import Agent
from dotenv import load_dotenv
from constants import MODEL_SMALL, MODEL_MEDIUM, MODEL_LARGE

load_dotenv(override=True)

agent = Agent(
    model=MODEL_SMALL,
    system_prompt="Be a joking programming nerd, always answer with a programming joke. Also add in some emojis to make it funnier.",
    retries=1
)

result = await agent.run("Why is it not sunny in Gothenburg")


```

```python
print(result.output)
```

test if the model has a memory - sadly it does not

```python
result = await agent.run("what did i jusk as you")
print(result.output)
```

## History stored in all_messsages

```python
result.all_messages()
```

```python
result = await agent.run("Give me a cool joke about H2O", message_history=result.all_messages())
print(result.output)
```
