# Prompt creation and versioning

```python
from mlflow.genai import register_prompt

email_summary_description = register_prompt(
    name="summary_description",
    template="summarize in 3-4 sentences.",
    commit_message="initial summary description prompt",
    tags={
        "author": "omeraytug", 
        "agent": "email_extractor", 
        "prompt_type": "description"})

email_summary_description
```

```python
email_summary_description = register_prompt(
    name="summary_description",
    template="summarize in {{num_sentences}} sentences, use same language as original text",
    commit_message="variable sentence and language specification",
    tags={
        "author": "omeraytug", 
        "agent": "email_extractor", 
        "prompt_type": "description"})

email_summary_description
```

## Load a prompt

```python
from mlflow.genai import load_prompt

email_summary_description = load_prompt("summary_description", version=2)
email_summary_description
```

```python
email_summary_description.format(num_sentences = 2)
```

```python
email_urgency_description = register_prompt(
    name="email-urgency-description",
    template="Assess urgency based on tone and content. "
    "'high' = deadlines, serious impact, angry tone; "
    "'medium' = inconvenience, no immediate deadline; "
    "'low' = general questions, minor issues.",
    commit_message="initial email urgency prompt",
    tags={
        "author": "kokchun",
        "agent": "email_extractor",
        "prompt_type": "description",
    },
)

email_extractor_system_prompt = register_prompt(
    name="email-extractor-system-prompt",
    template="You are an expert at extracting structured information from customer support emails. "
    "Extract the sender details, categorize the issue, assess urgency based on tone and content, "
    "and summarize the problem. ",
    commit_message="initial email extractor system prompt",
    tags={
        "author": "kokchun",
        "agent": "email_extractor",
        "prompt_type": "system prompt",
    },
)
```
